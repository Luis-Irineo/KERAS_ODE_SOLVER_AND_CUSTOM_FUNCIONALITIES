```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import SGD, RMSprop, Adam
import math
from tensorflow.keras.preprocessing import image

from matplotlib import pyplot as plt
from matplotlib.image import imread
import numpy as np
import os
```

```python
class GrayScale(tf.keras.layers.Layer):
    def __init__(self,**kwargs):
        super().__init__(**kwargs)
        self.threshold = 155
    
    def reduction(self, input):
        output = tf.reduce_sum(input * [0.2125, 0.7174, 0.0721],
                              axis = -1,
                              keepdims = True)
        return output
    def blk(self, input):
        gray = self.reduction(input)
        outputb = tf.where(gray > self.threshold, 1.0, 0.0)
        return outputb
        
    def call(self, input):
        output = self.reduction(input)
        return output
```

```python
os.chdir(r"/tf/keras_neural_network/Mis_Tests")
mi_imagen = "kr_W.jpg"
imtest = imread(mi_imagen)
print(imtest.shape)
plt.imshow(imtest)
plt.show()
```

```python
img = image.load_img(mi_imagen, target_size=(imtest.shape[0], imtest.shape[1]))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)  # Se agrega la dimensin de batch

grey_layer = GreyScale()
output = grey_layer(img_array)
output_image = output.numpy()[0]  # Se remueve la dimensin para visualizarla
print(output.shape)

plt.imshow(output_image, cmap = "gray")
plt.show()
```

```python
img = image.load_img(mi_imagen, target_size=(imtest.shape[0], imtest.shape[1]))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)  # Se agrega la dimensin de batch

black_layer = GrayScale()
output = black_layer.blk(img_array)
output_image = output.numpy()[0]  # Se remueve la dimensin para visualizarla
print(output.shape)

plt.imshow(output_image,cmap = "gray")
plt.show()
```
